```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
merge_df = pd.merge(city_df, ride_df, on="city", how="right")

# Display the data table for preview
merge_df
```

```python
#Y: average fare per city.
average_fare = merge_df.groupby("city")["fare"].mean()
average_fare
```

```python
#Z: total number of drivers per city.

total_drivers = merge_df.groupby("city")["driver_count"].mean()
total_drivers
```

```python
#city types
city_type  = merge_df.set_index('city')['type']
city_type.value_counts()
```

```python
#X: total rides per city.
total_rides = merge_df.groupby("city")["ride_id"].count()
total_rides
```

```python
#group by city
urban = merge_df.loc[(merge_df["type"] == "Urban")]
suburban = merge_df.loc[(merge_df["type"] == "Suburban")]
rural = merge_df.loc[(merge_df["type"] == "Rural")]
```

## Bubble Plot of Ride Sharing Data

```python


# Obtain the x and y coordinates for each of the three city types
plt.ylim(20, 45)

#x_axis = np.arrange(0,40,5)
plt.xlim(0, 40)

# Build the scatter plots for each city types
rural_bubble = plt.scatter(total_rides, average_fare, marker="o", facecolors="red", edgecolors="black",
            s=total_drivers*5,alpha = .75)
    
# Incorporate the other graph properties
plt.title('Pyber Ride Sharing Data(2016)')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare($)')

# Create a legend
plt.legend(loc='upper right')

# Incorporate a text label regarding circle size
```
